```python
import numpy as np
np.random.seed(42)
```

```python
a= np.random.multivariate_normal(mean=[0,0], cov= [[1.0, 0.5],[0.5,1.0]], size=100)
b= np.random.multivariate_normal(mean=[2.5,0], cov= [[1.2, -0.3],[-0.3,0.8]], size=100)

X= np.vstack((a,b))
y= np.array([0]*100 + [1]*100)
X.shape, y.shape, np.unique(y, return_counts=True)
```

```python
mu=X.mean(axis=0)
sigma=X.std(axis=0)
Z = (X - mu) / sigma
Z.mean(axis=0), Z.std(axis=0)
```

```python
w=np.array([0.9,-0.7])
b=0.1
s=Z.dot(w)+b
pred= (s>=0).astype(int)
accuracy= (pred==y).mean()
accuracy
tn= ((pred==0) & (y==0)).sum()
fn= ((pred==0) & (y==1)).sum()
tp= ((pred==1) & (y==1)).sum()
fp= ((pred==1) & (y==0)).sum()
confusion_matrix= np.array([[tn, fp],[fn, tp]])
confusion_matrix
```

```python
perm = np.random.permutation(len(Z))
Z = Z[perm]
y = y[perm]

n_train = int(0.7 * len(Z))
Z_train, Z_test = Z[:n_train], Z[n_train:]
y_train, y_test = y[:n_train], y[n_train:]

t_train = y_train * 2 - 1
t_test = y_test * 2 - 1
```

```python
phi_train = np.c_[np.ones(len(Z_train)), Z_train]
phi_test = np.c_[np.ones(len(Z_test)), Z_test]

w_hat = np.linalg.pinv(phi_train).dot(t_train)
print("Learned weights (bias, w1, w2)",w_hat)
```

```python
score_test = phi_test.dot(w_hat)
pred_test = (score_test >= 0).astype(int)

accuracy_test = (pred_test == y_test).mean()
print("Test set accuracy:", round(accuracy_test,3))
```
